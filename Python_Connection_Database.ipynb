**JOINS**

image.png

SQL Left (Outer) Join

```
query='select * from orders left join customers orders.customer_id=customers.customer_id'
data=pd.read_sql(query,conn)
data
```

SQL Right (Outer) Join

```
query='select * from orders right join customers orders.customer_id=customers.customer_id'
data=pd.read_sql(query,conn)
data
```

SQL Inner Join

```
query='select * from orders inner join customers orders.customer_id=customers.customer_id'
# or as below
query='select * from orders join customers orders.customer_id=customers.customer_id'
data=pd.read_sql(query,conn)
data
```
