**matplotlib -- plotting library**

```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
x,y=np.arange(0,10),np.arange(11,21)
```

**Scatter Plot**

```
plt.title('X Y Plot Graph in 2D')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.scatter(x,y,c='r')
plt.savefig('xy.png')
```

**plot**

```
plt.plot(x,y)
```

```
y=x*x
plt.plot(x,y)
```

```
plt.plot(x,y,'r*-')
```

```
plt.title('2d diagram')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x, y,'r*',linestyle='dashed',markersize=12)
```

```
plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'bo--')
```

```
x=np.arange(0,10)
y=3 * x +5
plt.title('Linear Function')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.plot(x,y,'r+--')
```

```
x=np.arange(0,5 * np.pi,0.1)
y=np.sin(x)
plt.title('Linear Function')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.plot(x,y,'b+--')
```

```
x=np.arange(0,5 * np.pi,0.1)
y=np.cos(x)
plt.title('Linear Function')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.plot(x,y,'r+--')
```

```
x=np.arange(0,4 * np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.title('Sine Function')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r+--')

plt.title('Cos Function')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
```

**Bar Plot**

```
x=[1,2,23]
y= [2,4,6]
x1=[3,4,6]
y1=[4,5,7]
plt.bar(x,y)
plt.bar(x1,y1,color='g')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

**Histograms**

```
a=np.array([22,23,4,43,56])
plt.hist(a,bins=10)
plt.title('histogram')
```

**Seaborn**

```
import seaborn as sns
```

```
df=sns.load_dataset("tips") # tip is dependent feature
df.head()
```

**Correlation with heatmap**

```
df.cov() # not required
```

```
df.corr() # if positive indicted direction of relationship and strength by magnitude
```

```
sns.heatmap(df.corr())
```

```
# Bi variate Analysis
sns.jointplot('total_bill','tip',data=df,kind='hex') 
```

```
sns.jointplot('total_bill','tip',data=df,kind='reg')
```

```
sns.pairplot(df)
```

```
sns.pairplot(df, hue='sex')
```

```
sns.pairplot(df, hue='smoker')
```

```
df.smoker.value_counts()
```

```
sns.distplot(df['tip'], kde=False, bins=10) # kernel density estimation
```

```
sns.distplot(df['tip'], kde=True, bins=10) # kernel density estimation
```

```
sns.countplot('sex',data=df)
```

```
sns.countplot(y='smoker',data=df) # x and y are optional
```

```
sns.barplot(x='sex',y='total_bill', data=df)
```

```
sns.boxplot('smoker','total_bill',data=df)
```
